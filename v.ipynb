```python
import pandas as pd
import plotly.express as px

import numpy as np
```

```python
ANSWER: float = 0.25 * np.pi + 1.25 * np.asin(0.8) - 1
```

```python
df = pd.read_csv("monte_carlo_results_large.csv")
df = df.rename(columns={
    "wide_area": "Большая область",
    "narrow_area": "Маленькая область",
    "large_area": "Очень большая область",
    "wide_relative_error": "Ошибка (Большая область)",
    "narrow_relative_error": "Ошибка (Маленькая область)",
    "large_relative_error": "Ошибка (Очень большая область)"
})
df.head()
```

```python
fig = px.line(
    df,
    x="N", y=["Маленькая область", "Большая область", "Очень большая область"],
    title="Оценка площади пересечание кругов",
    labels={"variable": ""}
)
fig.add_hline(
    ANSWER,
    annotation_text="Точное значение",
    line_dash="dash",
    line_color="black",
)
fig.update_layout(
    xaxis_title="Количество точек",
    yaxis_title="Площадь пересечения"
)
```

```python
fig = px.line(
    df,
    x="N", y=["Ошибка (Маленькая область)", "Ошибка (Большая область)", "Ошибка (Очень большая область)"],
    title="Относительная ошибка при подсчете пересечание кругов",
    labels={"variable": ""}
)
    
fig.update_layout(
    xaxis_title="Количество точек",
    yaxis_title="Относительная ошибка"
)
```
